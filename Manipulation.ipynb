```python
##Importing and merging datasets
pop = pd.read_csv("census-populations.csv")
region = pd.read_csv("census-regions.csv")
census = pd.merge(pop, region)
census.head()
```

```python
##Re-shape census so that one column contains all population measures, and another the year attributes.
census_reshaped = pd.melt(frame = census, id_vars = ['state', 'region', 'division'], value_vars = ['2010', '2011', '2012', '2013', '2014', '2015', '2016'],var_name = 'Year', value_name = 'Population')
census_reshaped
```

```python
#Group the data by year and summarize it.
census_reshaped.groupby(['Year']).mean()
```

```python
#Group the data by region and summarize it. 
census_reshaped.groupby(['region']).mean()
```

```python
#Group the data by division and summarize it. 
census_reshaped.groupby(['division']).mean()
```

```python
#Group the data by year and summarize it. 
census_reshaped.groupby(['Year', 'region', 'division']).mean()
```
